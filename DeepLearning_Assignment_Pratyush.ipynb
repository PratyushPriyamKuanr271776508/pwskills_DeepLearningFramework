## TensorFlow Questions

### 1. What is TensorFlow 2.0, and how is it different from TensorFlow 1.x?
TensorFlow 2.0 is the updated version of Google's open-source deep learning framework. Key differences from TensorFlow 1.x include:
- Eager execution is enabled by default (vs. graph execution in 1.x)
- Simplified API with Keras integration as the high-level API
- Removal of redundant APIs and more intuitive function organization
- Better support for deployment on various platforms
- Improved performance and easier debugging

### 2. How do you install TensorFlow 2.0?
You can install TensorFlow 2.0 using pip:
```bash
pip install tensorflow
```
For GPU support:
```bash
pip install tensorflow-gpu
```

### 3. What is the primary function of the tf.function in TensorFlow 2.0?
`tf.function` is a decorator that converts Python functions into TensorFlow graphs. This provides:
- Performance optimization through graph execution
- Portability (saved models can be loaded without Python code)
- Better performance for computationally intensive operations

### 4. What is the purpose of the Model class in TensorFlow 2.0?
The Model class in TensorFlow 2.0 (typically `tf.keras.Model`) is the base class for defining custom neural network models. It provides:
- Methods for training (`fit()`), evaluation (`evaluate()`), and prediction (`predict()`)
- Infrastructure for layer organization and connectivity
- Weight management and saving/loading capabilities
- Custom training loop support

### 5. How do you create a neural network using TensorFlow 2.0?
Here's a basic example using the Keras API:
```python
import tensorflow as tf
from tensorflow.keras import layers

model = tf.keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

### 6. What is the importance of Tensor Space in TensorFlow?
Tensor Space refers to the n-dimensional space where tensors (multi-dimensional arrays) operate. Its importance includes:
- Provides the mathematical foundation for neural network operations
- Enables efficient computation through vectorized operations
- Allows visualization of high-dimensional data through dimensionality reduction
- Forms the basis for tensor operations like dot products, convolutions, etc.

TensorFlow Playground is an interactive web application that:

### 9. What is Netron, and how is it useful for deep learning models?
Netron is a viewer for neural network models that:

### 10. What is the difference between TensorFlow and PyTorch?
Key differences:
- **Eager execution**: PyTorch is eager by default, TensorFlow 2.0 adopted this
- **API design**: PyTorch is more Pythonic, TensorFlow has broader production features
- **Static vs dynamic graphs**: TensorFlow uses static graphs (via tf.function), PyTorch uses dynamic computation graphs
- **Deployment**: TensorFlow has better mobile/embedded support
- **Community**: PyTorch is more popular in research, TensorFlow in production
- **Visualization**: TensorFlow has better built-in tools (TensorBoard)

### 12. What is the basic structure of a PyTorch neural network?

```python

class Net(nn.Module):
    def __init__(self):

        self.fc1 = nn.Linear(784, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, 10)
    
    def forward(self, x):
```

### 16. What are some common activation functions used in neural networks?
Common activation functions:
- ReLU (Rectified Linear Unit): `f(x) = max(0, x)`
- Sigmoid: `f(x) = 1 / (1 + exp(-x))`
- Tanh (Hyperbolic Tangent): `f(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x))`
- Leaky ReLU: `f(x) = max(0.01x, x)`
- Softmax (for output layers in classification)
- Swish: `f(x) = x * sigmoid(x)`

### 17. What is the difference between torch.nn.Module and torch.nn.Sequential in PyTorch?
- `nn.Module`: Base class for all neural network modules, allows custom model definitions with full flexibility
- `nn.Sequential`: Container module that allows simple sequential stacking of layers (less flexible but more concise)
Example of Sequential:
```python
model = nn.Sequential(
    nn.Linear(784, 64),
    nn.ReLU(),
    nn.Linear(64, 10)
)
```
